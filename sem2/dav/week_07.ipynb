# Week 7 - Matplotlib & Seaborn Visualizations: MTCARS & IPL Matches

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## MTCARS Visualizations

```python
df = pd.read_csv('mtcars.csv')
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', None)
df
```

```python
# 1. Line plot of MPG values
mpg = df['mpg']
mpg.plot()
plt.title('MPG Values')
plt.show()
```

```python
# 2. Line plot of cyl, wt, and mpg
df_subset = df[['cyl', 'wt', 'mpg']]
df_subset.plot()
plt.title('Cyl, Wt, and MPG')
plt.show()
```

```python
# 3. Bar plot: count of cars by number of cylinders
cyl_counts = df['cyl'].value_counts()
cyl_counts.plot(kind='bar', color='green')
plt.title('Count of Cars by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Scatter plot: weight vs MPG
df.plot(kind='scatter', x='wt', y='mpg', figsize=(9, 9), color='black')
plt.title('Weight vs MPG')
plt.show()
```

```python
# 5. Box plot: MPG for each cylinder category
plt.figure(figsize=(8, 6))
sns.boxplot(x='cyl', y='mpg', data=df)
plt.title('MPG by Cylinder Category')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
# 6. Histogram of MPG
x = df['mpg']
plt.hist(x)
plt.title('Distribution of MPG in Cars Dataset')
plt.ylabel('Frequency')
plt.xlabel('MPG')
plt.show()
```

```python
# 7. Pie chart: transmission types
df2 = pd.read_csv('mtcars.csv')
am_counts = df2['am'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(am_counts, labels=['Automatic', 'Manual'], autopct='%1.1f%%',
        colors=['gold', 'lightcoral'], startangle=90)
plt.title('Proportion of Cars by Transmission Type')
plt.show()
```

```python
# 8. Nested pie chart: cylinders (outer) and gears (inner)
outer_labels = df2['cyl'].value_counts().index
outer_sizes = df2['cyl'].value_counts().values
inner_labels = df2['gear'].value_counts().index
inner_sizes = df2['gear'].value_counts().values

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(outer_sizes, labels=outer_labels, radius=1, autopct='%1.1f%%',
       colors=['skyblue', 'lightgreen', 'salmon'])
ax.pie(inner_sizes, labels=inner_labels, radius=0.7, autopct='%1.1f%%',
       colors=['pink', 'plum', 'peachpuff'])
ax.set(aspect='equal')
plt.title('Nested Pie Chart: Cylinders (Outer) and Gears (Inner)')
plt.show()
```

## IPL Matches Analysis

```python
df = pd.read_csv('matches.csv')
```

```python
# 1. Names of IPL teams
teams = set(df['team1']).union(set(df['team2']))
teams = list(teams)
print(teams)
```

```python
# 2. Total matches won by each team
wins = df['winner'].value_counts()
wins.plot(kind='bar', title='Number of matches won by each team',
          color='blue', xlabel='Teams', ylabel='Number of matches')
plt.tight_layout()
plt.show()
```

```python
# 3. Umpires as second umpire and match count
print(df['umpire2'].value_counts())
```

```python
# 4. Matches held in each venue
venue = df['venue'].value_counts()
plt.figure(figsize=(12, 8))
venue.plot(kind='bar', title='Number of matches held in each venue',
           color='red', width=0.7, xlabel='Venues', ylabel='Number of matches')
plt.tight_layout()
plt.show()
```

```python
# 5. Top 10 players of the match
top = df['player_of_match'].value_counts().head(10)
plt.figure(figsize=(15, 9))
sns.barplot(x=top.index, y=top.values)
plt.title('Top 10 players of the match')
plt.xlabel('Players')
plt.ylabel('Number of matches')
plt.show()
```

```python
# 6. Bottom 10 venues by match count
ven = df['venue'].value_counts().tail(10)
sns.barplot(x=ven.index, y=ven.values)
plt.title('Bottom 10 venues where the least number of matches were played')
plt.xlabel('Venues')
plt.ylabel('Number of matches')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 7. Wins at MA Chidambaram Stadium, Chepauk
venues = df[df['venue'] == 'MA Chidambaram Stadium, Chepauk']
wins = venues['winner'].value_counts()
sns.barplot(x=wins.index, y=wins.values)
plt.title('Number of Wins of Each Team at MA Chidambaram Stadium, Chepauk', fontsize=16)
plt.xlabel('Teams')
plt.ylabel('Wins')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 8. Venues where matches ended in a tie
ties = df[df['result'] == 'tie']
venues = ties['venue']
print(venues)
```

```python
# 9. Top 5 teams season-wise (heatmap)
top = df.groupby(['season', 'winner']).size().reset_index(name='wins')
pivot_table = top.pivot_table(index='season', columns='winner', values='wins', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='g')
plt.title('Top 5 Teams Season-wise by Number of Wins (Heatmap)')
plt.xlabel('Teams')
plt.ylabel('Season')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 10. Relationship between winning toss and winning the match
df['toss_win_match_win'] = df['toss_winner'] == df['winner']
toss_win_match_win_count = df['toss_win_match_win'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=toss_win_match_win_count.index, y=toss_win_match_win_count.values)
plt.title('Relationship between Winning the Toss and Winning the Match')
plt.xlabel('Toss Winner Wins the Match?')
plt.ylabel('Number of Matches')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
print(toss_win_match_win_count)
```
