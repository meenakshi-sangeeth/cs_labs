## Lab Manual

```python
# 1. Generate a 3x4 NumPy array with random integers between 1 and 50
# a. Mean, Median, Standard Deviation
# b. Sum of all elements and sum of each row
# c. Reshape to 2x6

a = np.random.randint(1, 50, 12).reshape(3, 4)
print('Matrix: ')
print(a)
print('Mean: ', np.mean(a))
print('Median: ', np.median(a))
print('Standard Deviation', np.std(a))
print('Sum of all elements: ', np.sum(a))
x = np.sum(a, axis=1)
for i in range(3):
    print('Sum of row', i + 1, ':', x[i])
print('2x6 matrix: ')
print(a.reshape(2, 6))
```

```python
# 2. Create two 3x3 matrices and perform matrix operations
# a. Addition  b. Subtraction  c. Multiplication & dot product
# d. Transpose  e. Determinant and inverse

a1 = np.random.randint(11, 20, 9).reshape(3, 3)
print('Matrix 1:\n', a1)
a2 = np.random.randint(1, 10, 9).reshape(3, 3)
print('Matrix 2:\n', a2)

print('Matrix sum:')
print(a1 + a2)
print('Matrix difference:')
print(a1 - a2)
print('Matrix multiplication:')
print(a1 * a2)
print('Dot product:')
print(np.dot(a1, a2))
print('Transpose of matrix 1: ')
print(a1.T)
print('Transpose of matrix 2: ')
print(a2.T)
print('Determinant of matrix 1:', np.linalg.det(a1))
print('Determinant of matrix 2:', np.linalg.det(a2))
print('Inverse of matrix 1:\n', np.linalg.inv(a1))
print('Inverse of matrix 2:\n', np.linalg.inv(a2))
```

## Additional Exercises

```python
# 1. Given arr = np.arange(1, 101).reshape(10, 10)
# a) First 3 rows and last 3 columns
# b) Diagonal elements

arr = np.arange(1, 101).reshape(10, 10)
print('First 3 rows:')
print(arr[:3])
print('Last 3 columns:')
print(arr[:, -3:])
print('Diagonal elements: ')
print(np.diag(arr))
```

```python
# 2. Create a 3D array of shape (4, 5, 6) with random integers 1-100
# Extract from second row of every 2D matrix along first axis
# Extract from last column of the third 2D matrix

a = np.random.randint(1, 100, 120).reshape(4, 5, 6)
print('Matrix:\n', a)
print('\nElements from the second row of every 2D matrix along the first axis:')
print(a[:, 1:2, :1])
print('\nElements from the last column of the third 2D matrix along the first axis:')
print(a[2:3, 0:1, 5:])
```

```python
# 3. Create a 4D array of shape (3, 4, 5, 6) with random floats
# Extract alternate rows and columns from second and fourth dimensions
# Extract every 2nd matrix along first axis and every 3rd along third

a = np.random.rand(3, 4, 5, 6)
print('Matrix:\n', a)
print('\nAlternate row and column from the second and fourth dimensions:')
print(a[:, ::2, :, ::2])
print('Elements from every 2nd matrix along the first axis and every 3rd along third:')
print(a[::2, :, ::3, :])
```

```python
# 4. Create a 2D array of shape (6, 8) with numbers from 1 to 48
# Reverse the rows; reverse the last three columns

a = np.arange(1, 49).reshape(6, 8)
print('Matrix:\n', a)
rev = a[::-1]
print('\nMatrix after reversing rows:')
print(rev)
print('\nMatrix after reversing the last three columns:')
a[:, -3:] = a[:, -3:][:, ::-1]
print(a)
```

```python
# 5. Create a 3D array of shape (3, 5, 5) with random integers 10-100
# Extract multiples of 3; replace elements divisible by 5 with 0

a = np.random.randint(10, 100, 75).reshape(3, 5, 5)
print('Matrix:\n', a)
print('\nElements which are multiples of 3:\n', a[a % 3 == 0])
a[a % 5 == 0] = 0
print('\nMatrix after replacing elements divisible by 5:\n', a)
```
