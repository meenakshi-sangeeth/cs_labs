# Week 8 - IPL Deliveries Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('deliveries.csv')
df
```

```python
# 1. Total runs scored by each team
total_runs = df.groupby('batting_team')['total_runs'].sum().reset_index()
print(total_runs)
plt.figure(figsize=(12, 6))
sns.barplot(data=total_runs, x='batting_team', y='total_runs', palette='copper')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.tight_layout()
plt.show()
```

```python
# 2. Top 10 highest scoring batters
top_batters = df.groupby('batter')['batsman_runs'].sum().nlargest(10)
print(top_batters)
```

```python
# 3. Top 10 bowlers with most wickets
df['is_wicket'] = df['is_wicket'].astype(bool)
top_bowlers = df[df['is_wicket']].groupby('bowler')['is_wicket'].sum().nlargest(10)
print(top_bowlers)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_bowlers.index, y=top_bowlers.values, palette='magma')
plt.title('Top 10 Bowlers with Most Wickets')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. Total extras conceded by each team
total_extras_team = df.groupby('bowling_team')['extra_runs'].sum().reset_index()
print(total_extras_team)
plt.figure(figsize=(12, 6))
sns.barplot(data=total_extras_team, x='bowling_team', y='extra_runs', palette='mako')
plt.title('Total Number of Extras Conceded by Each Team')
plt.xlabel('Team')
plt.ylabel('Extra Runs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 5. Top 5 batsmen involved in most run-outs
run_outs = df[df['dismissal_kind'] == 'run out']['batter'].value_counts().head(5)
print(run_outs)
plt.figure(figsize=(12, 6))
sns.barplot(x=run_outs.index, y=run_outs.values, palette='crest')
plt.title('Top 5 Batsmen Involved in Most Run-outs')
plt.xlabel('Batter')
plt.ylabel('Run-outs')
plt.show()
```

```python
# 6. Top 5 bowlers conceding most extras
bowlers_extras = df.groupby('bowler')['extra_runs'].sum().nlargest(5)
print(bowlers_extras)
plt.figure(figsize=(12, 6))
sns.barplot(x=bowlers_extras.index, y=bowlers_extras.values, palette='YlOrRd')
plt.title('Top 5 Bowlers Conceding Most Extras')
plt.xlabel('Bowler')
plt.ylabel('Extras')
plt.show()
```

```python
# 7. Most common dismissal type (pie chart)
dismisal_kinds = df['dismissal_kind'].value_counts().reset_index()
dismisal_kinds.columns = ['dismissal_kind', 'count']
plt.figure(figsize=(10, 10))
palette_color = sns.color_palette('twilight_shifted_r')
plt.pie(dism_kinds['count'], colors=palette_color, autopct='%1.0f%%')
plt.legend(dism_kinds['dismissal_kind'], title='Dismissal Kinds', loc='best')
plt.title('Most Common Dismissal Types')
plt.show()
```

```python
# 8. Top 3 overs with highest run rate
overs_runrate = df.groupby('over')['total_runs'].sum() / df['over'].value_counts().sort_index()
top_overs = overs_runrate.nlargest(3)
print(top_overs)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_overs.index, y=top_overs.values, palette='cividis')
plt.title('Top 3 Overs with Highest Run Rate')
plt.xlabel('Over')
plt.ylabel('Run Rate')
plt.show()
```

```python
# 9. Powerplay performance (first 6 overs)
powerplay_runs = df[df['over'] <= 6].groupby('batting_team')['total_runs'].sum()
powerplay_runs.plot(kind='bar', figsize=(12, 6), color='maroon',
                    title='Total Runs Scored in Powerplay')
plt.xlabel('Team')
plt.ylabel('Runs')
plt.tight_layout()
plt.show()
```

```python
# 10. Top 5 batsmen with most sixes and fours
sixes = df[df['batsman_runs'] == 6].groupby('batter')['batsman_runs'].count().nlargest(5)
fours = df[df['batsman_runs'] == 4].groupby('batter')['batsman_runs'].count().nlargest(5)
plt.figure(figsize=(12, 6))
sns.barplot(x=sixes.index, y=sixes.values, color='blue', label='Sixes')
sns.barplot(x=fours.index, y=fours.values, color='maroon', label='Fours')
plt.legend()
plt.title('Top 5 Batsmen with Most Sixes and Fours')
plt.xlabel('Batter')
plt.ylabel('Count')
plt.show()
```
