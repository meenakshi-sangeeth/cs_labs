# Week 6 - Exploratory Data Analysis: MTCARS & Cereals

```python
import pandas as pd
import matplotlib.pyplot as plt

dat_mtcars = pd.read_csv('mtcars.csv', index_col=0)
dat_mtcars
```

## MTCARS Analysis

```python
# 1. Distribution of fuel efficiency
dat_mtcars['mpg'].plot(kind='hist', bins=10, title='Histogram of the Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

dat_mtcars['mpg'].plot(kind='box')
plt.ylabel('MPG')
plt.show()
```

```python
# 2. Cars with best and worst fuel efficiency
max_mpg = dat_mtcars['mpg'].max()
print('Best fuel efficiency:')
print(dat_mtcars[dat_mtcars['mpg'] == max_mpg])

min_mpg = dat_mtcars['mpg'].min()
print('Worst fuel efficiency:')
print(dat_mtcars[dat_mtcars['mpg'] == min_mpg])
```

```python
# 3. Manual vs automatic count
print('Number of transmission types:', dat_mtcars['am'].nunique())
print(dat_mtcars['am'].value_counts())
```

```python
# 4. Car with worst horsepower
min_hp = dat_mtcars['hp'].min()
print('Car with worst horsepower:\n')
print(dat_mtcars[dat_mtcars['hp'] == min_hp]['hp'])
```

```python
# 5. 5-number summary and boxplot of displacement
summary = dat_mtcars.describe().loc[['min', '25%', '50%', '75%', 'max']]
print('Five number summary:\n')
print(summary)
dat_mtcars['disp'].plot(kind='box', title='Boxplot of displacement')
plt.show()
```

```python
# 6. Heaviest car and its gear count
wt_max = dat_mtcars['wt'].max()
print('Heaviest car:\n')
print(dat_mtcars[dat_mtcars['wt'] == wt_max]['gear'])
```

```python
# 7. Car with best qsec
qsec_max = dat_mtcars['qsec'].max()
print('Car with the best qsec:\n')
print(dat_mtcars[dat_mtcars['qsec'] == qsec_max]['qsec'])
```

```python
# 8. Average MPG for manual vs automatic cars
dat_man = dat_mtcars[dat_mtcars['am'] == 1]['mpg']
print('Average MPG of manual cars:', dat_man.mean())
dat_auto = dat_mtcars[dat_mtcars['am'] == 0]['mpg']
print('Average MPG of automatic cars:', dat_auto.mean())
```

```python
# 9. Side-by-side boxplot: Manual vs Automatic fuel efficiency
x = dat_mtcars.plot(kind='box', column='mpg', by='am', title='Manual vs Automatic')
x[0].set_xlabel('Automatic: 0\nManual: 1')
x[0].set_ylabel('MPG')
plt.show()
```

```python
# 10. Relationship between weight and MPG
dat_x = dat_mtcars[['mpg', 'wt']].reset_index(drop=True)
dat_x.plot(kind='scatter', x='mpg', y='wt', color='green', title='Weight vs MPG')
plt.show()
```

```python
# 11. Cars by gear count
print('Number of gear types:', dat_mtcars['gear'].nunique())
print(dat_mtcars['gear'].value_counts())
```

```python
# 12. Fuel efficiency vs number of gears
dat_x = dat_mtcars[['mpg', 'gear']]
dat_x.plot(kind='scatter', x='mpg', y='gear', color='green', title='MPG vs Gears')
plt.show()
```

```python
# 13. Horsepower vs number of cylinders
dat_x = dat_mtcars[['hp', 'cyl']]
dat_x.plot(kind='scatter', x='hp', y='cyl', color='orange', title='HP vs Cylinders')
plt.show()
```

```python
# 14. Displacement vs gross horsepower
dat_x = dat_mtcars[['disp', 'hp']]
dat_x.plot(kind='scatter', x='disp', y='hp', color='violet', title='Displacement vs HP')
plt.show()
```

```python
# 15. Best car for high speed and good fuel efficiency
df_sorted = dat_mtcars.sort_values(by=['hp', 'mpg'], ascending=[False, False])
print(df_sorted.iloc[0])
```

## Cereals Analysis

```python
cdf = pd.read_csv('Cereals1.csv')
cdf = pd.DataFrame(cdf)
n = cdf.select_dtypes(include='number').columns
for i in n:
    mean_value = cdf[i].mean()
    cdf[i].replace(-1, mean_value, inplace=True)
```

```python
# 1. Unique brands
print('Number of unique brands:', cdf['name'].nunique())
```

```python
# 2. Cereals per manufacturer
print('Cereals per manufacturer:', cdf['mfr'].value_counts(), sep='\n')
```

```python
# 3. Hot vs cold cereals
print('Cold(C) and Hot(H) cereals:', cdf['type'].value_counts(), sep='\n')
```

```python
# 4. Best and worst cereal by rating
r_max = cdf['rating'].max()
print('Best cereal:')
print(cdf[cdf['rating'] == r_max]['name'])
r_min = cdf['rating'].min()
print('\nWorst cereal:')
print(cdf[cdf['rating'] == r_min]['name'])
```

```python
# 5. Hot vs cold rating comparison
rating = cdf.groupby('type')['rating'].mean()
print(rating)
print('\nMost preferred:', rating.idxmax())
```

```python
# 6. Highest fiber and lowest sugar
fib = cdf['fiber'].max()
print('Cereals with highest fiber:')
print(cdf[cdf['fiber'] == fib]['name'])
sug = cdf['sugars'].min()
print('\nCereals with lowest sugar:')
print(cdf[cdf['sugars'] == sug]['name'])
```

```python
# 7. Cereals with more than 3g protein
p = cdf[cdf['protein'] > 3]
print('Cereals with more than 3 grams of protein:')
print(p[['name', 'protein']])
```

```python
# 8. Cereals by display shelf
shelf_count = cdf['shelf'].value_counts()
print('\nCereals by display shelf:')
print(shelf_count)
```

```python
# 9. Average sugar per brand
sugar = cdf.groupby('name')['sugars'].mean()
print('Average sugar content across different brands:')
print(sugar)
```

```python
# 10. Average calories per manufacturer
avg = cdf.groupby('mfr')['calories'].mean()
print('Average calories per manufacturer:')
print(avg)
```

```python
# 11. Average nutritional content
avg_nutrition = cdf[['calories', 'sugars', 'protein', 'fiber']].mean()
print('\nAverage nutritional content of cereals:')
print(avg_nutrition)
```

```python
# 12. Scatter: sugars vs calories
dat_x = cdf[['sugars', 'calories']]
dat_x.plot(kind='scatter', x='sugars', y='calories', color='violet', title='Sugars vs Calories')
plt.show()
```

```python
# 13. Sugar content in high-rated vs low-rated cereals
high_rating_sugars = cdf[cdf['rating'] > cdf['rating'].median()]['sugars']
low_rating_sugars = cdf[cdf['rating'] <= cdf['rating'].median()]['sugars']
print('\nSugar content in cereals with high ratings:')
print(high_rating_sugars.describe())
print('\nSugar content in cereals with low ratings:')
print(low_rating_sugars.describe())
```

```python
# 14. Healthy cereals (fiber > 3, sugars < 5)
healthy_cereals = cdf[(cdf['fiber'] > 3) & (cdf['sugars'] < 5)]
print('\nHealthy cereals with ratings:')
print(healthy_cereals[['name', 'rating', 'fiber', 'sugars']])
```

```python
# 15. Scatter: rating vs shelf placement
dat_x = cdf[['rating', 'shelf']]
dat_x.plot(kind='scatter', x='rating', y='shelf', color='orange', title='Rating vs Shelf')
plt.show()
```
