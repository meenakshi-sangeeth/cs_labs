# Week 9 - Sentiment Analysis on Social Media Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sentiment.csv')
df
```

```python
# 1. Trim whitespace and lowercase Sentiment and Text columns
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Text'] = df['Text'].str.strip().str.lower()
df
```

```python
# 2. Sentiment categories and counts
sentiment_counts = df['Sentiment'].value_counts()
print('\nSentiment Categories and Counts:')
print(sentiment_counts.head(50))

plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.head(50).index, y=sentiment_counts.head(50).values,
            palette='Greens_r')
plt.title('Distribution of Sentiment Categories')
plt.xlabel('Sentiment')
plt.ylabel('Number of Texts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 3. Text length per sentiment category
df['Text_Length'] = df['Text'].apply(len)
sentiment_length = df.groupby('Sentiment')['Text_Length'].mean().head(25).sort_values(ascending=False)
print('\nAverage Text Length by Sentiment (Top 25):')
print(sentiment_length)

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Sentiment', y='Text_Length', order=sentiment_length.index)
plt.title('Text Length Distribution by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Text Length')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 4. Text(s) with maximum number of likes
max_likes = df['Likes'].max()
top_liked_texts = df[df['Likes'] == max_likes][['Text', 'Sentiment']]
print(f'\nText(s) with Maximum Likes ({max_likes}):')
top_liked_texts
```

```python
# 5. Text(s) with maximum reposts
max_reposts = df['Retweets'].max()
top_reposted_texts = df[df['Retweets'] == max_reposts][['Text', 'Sentiment']]
print(f'\nText(s) with Maximum Reposts ({max_reposts} reposts):')
print(top_reposted_texts)
```

```python
# 6. Top 5 countries with most positive sentiment texts
positive_countries = df[df['Sentiment'] == 'positive'].groupby('Country').size()
top_positive_countries = positive_countries.sort_values(ascending=False).head(5)
print('\nTop 5 Countries with Most Positive Sentiment Texts:')
print(top_positive_countries)

plt.figure(figsize=(10, 6))
top_positive_countries.plot(kind='bar')
plt.title('Top 5 Countries with Most Positive Sentiment Texts')
plt.xlabel('Country')
plt.ylabel('Number of Positive Texts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7. Texts with 'Travel' in hashtags
travel_texts = df[df['Hashtags'].str.contains('travel', case=False, na=False)][['Text']]
print("\nTexts with 'Travel' in Hashtags:")
print(travel_texts)
```

```python
# 8. Top 5 most liked negative sentiment posts from 2023
negative_2023 = df[(df['Year'] == 2023) & (df['Sentiment'] == 'negative')]
top_negative_2023 = negative_2023.nlargest(5, 'Likes')[['Text', 'Likes', 'User', 'Timestamp']]
print('\nTop 5 Most Liked Negative Sentiment Posts from 2023:')
print(top_negative_2023)
```

```python
# 9. Users with 'Happiness' sentiment posts
happiness_posts = df[df['Sentiment'] == 'happiness'][['User', 'Text', 'Likes', 'Timestamp']]
print("\nUsers and Posts with 'Happiness' Sentiment:")
print(happiness_posts)
```

```python
# 10. Sentiment trends by year and country (positive, negative, neutral)
selected_sentiments = ['positive', 'negative', 'neutral']
filtered_df = df[df['Sentiment'].isin(selected_sentiments)]
sentiment_by_year_country = filtered_df.groupby(['Year', 'Country', 'Sentiment']).size().unstack(fill_value=0)
print(sentiment_by_year_country)

sentiment_by_year_country.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Sentiment Trends by Year and Country')
plt.xlabel('Year and Country')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
